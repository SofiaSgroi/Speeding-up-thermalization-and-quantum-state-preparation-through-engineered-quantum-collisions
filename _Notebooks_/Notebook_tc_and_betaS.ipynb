```python
import os

fpath = os.getcwd()[:-12]+"\\Fixed_Length_GA"
print(fpath)
os.chdir(fpath)
os.getcwd()

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#T = 3
tc_list = [1.5, 1, 0.75, 0.5, 0.25, 0.125, 0.1]
Nc_list = [2, 3, 4, 6, 12, 24, 30]

results_beta1 = [-np.loadtxt("results/genericAncillas_n0tobeta1_tc" + str(tc_list[i]).replace(".", "") + "Nc" + str(Nc_list[i]) +"/history")[-1] for i in range(len(tc_list))]
results_beta05 = [-np.loadtxt("results/genericAncillas_n0tobeta05_tc" + str(tc_list[i]).replace(".", "") + "Nc" + str(Nc_list[i]) +"/history")[-1] for i in range(len(tc_list))]
```

```python
plt.plot(tc_list, results_beta1, marker = "s", markersize=10, label = r"$\beta = 1$", color="#0072b2")
plt.plot(tc_list, results_beta05, marker = "o", markersize=10, label = r"$\beta = 0.5$", color="#cc79a7")
plt.legend(fontsize=12)


plt.xlabel(r"$t_c$", fontsize = 14)
plt.ylabel(r"$-J$", fontsize = 14)

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

#plt.savefig("figure_thermalization3.pdf", bbox_inches="tight")
```
