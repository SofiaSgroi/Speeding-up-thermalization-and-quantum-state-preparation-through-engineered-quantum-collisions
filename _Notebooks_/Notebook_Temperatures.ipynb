```python
N_levs = 20
Omega_S = 1.
g = 1.
beta_range = 5.
npdtype ="complex128"

def temperatures(t_c, N_c):
    string = np.loadtxt("results/thermalAncillas_n0tobeta1_tc" + str(t_c).replace(".", "") + "Nc" + str(N_c) + "/solution")
    return t_c*np.array([n_c for n_c in range(1, N_c+1)]), (-beta_range +2.*beta_range*string)
```

```python
x1, y1 = temperatures(1, 1)
plt.plot(x1, y1, 'o', label="$t_c=1$")

x05, y05 = temperatures(0.5, 2)
plt.plot(x05, y05, 'o', label="$t_c=0.5$")

x01, y01 = temperatures(0.1, 10)
plt.plot(x01, y01, 'o', label="$t_c=0.1$")

plt.legend()
```
